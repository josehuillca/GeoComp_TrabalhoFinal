```python
coord_x = [10,160,280,320,380,360,340,380,325,280,220,150,80,30,80,30,80,10,70]
coord_y = [105,75,70,120,160,210,270,280,290,380,325,340,310,280,260,235,160,140]
#points = list(zip(coord_x, coord_y))
points = [Point(*coord) for coord in zip(coord_x, coord_y)]

a = QTree(1, 400,400, points)
a.subdivide()
a.draw()
```
